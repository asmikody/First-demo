```python
import pandas as pd
df = pd.read_csv(r"C:\Users\BAPS\Downloads\medical.csv")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset
# (Assume the dataset is named 'insurance.csv' with columns: age, sex, bmi, children, smoker, region, charges)
df = pd.read_csv('')

# Step 2: Bar Plot – Average medical charges between smokers and non-smokers
avg_charges = df.groupby('smoker')['charges'].mean().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(x='smoker', y='charges', data=avg_charges, palette='coolwarm')
plt.title('Average Medical Charges: Smokers vs Non-Smokers')
plt.xlabel('Smoker Status')
plt.ylabel('Average Charges')
plt.show()

# Step 3: Box Plot – Distribution of charges across genders
plt.figure(figsize=(6,4))
sns.boxplot(x='sex', y='charges', data=df, palette='Set2')
plt.title('Distribution of Medical Charges by Gender')
plt.xlabel('Gender')
plt.ylabel('Charges')
plt.show()

# Step 4: Scatter Plot – BMI vs Charges colored by Smoker Status
plt.figure(figsize=(8,6))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df, palette='husl', alpha=0.7)
plt.title('BMI vs Medical Charges (Colored by Smoker Status)')
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Medical Charges')
plt.show()
```
