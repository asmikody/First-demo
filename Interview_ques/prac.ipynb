```python
import re   # for email validation

class User:
    def __init__(self, username, password, email):
        # private attributes
        self.__username = username
        self.__password = None
        self.__email = None
        
        # use setters to validate on creation
        self.set_password(password)
        self.set_email(email)

    # ---------- SETTERS ----------
    def set_password(self, password):
        if len(password) < 8:
            print("Error: Password must be at least 8 characters long.")
        else:
            self.__password = password

    def set_email(self, email):
        if User.validate_email(email):
            self.__email = email
        else:
            print("Error: Invalid email format.")

    # ---------- GETTERS ----------
    def get_username(self):
        return self.__username

    def get_email(self):
        return self.__email

    # ---------- STATIC METHOD ----------
    @staticmethod
    def validate_email(email):
        pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
        return re.match(pattern, email) is not None


# --- Example Usage ---
user1 = User("asmi", "pass123", "asmi@gmail.com")
user2 = User("rahul", "password123", "rahulmail.com")

print("\nUser:", user1.get_username())
print("Email:", user1.get_email())
```

```python
'''Que 8: Implement a Plugin System Create a base abstract class Plugin with an abstract method run().
- Then, create different plugins like PrintPlugin, MathPlugin, etc., each overriding run().
- Create a PluginManager to load and execute all plugins.
 '''

from abc import ABC, abstractmethod

# ---------- Base Abstract Class ----------
class Plugin(ABC):
    @abstractmethod
    def run(self):
        pass  # abstract method must be implemented by subclasses


# ---------- Plugin Implementations ----------
class PrintPlugin(Plugin):
    def run(self):
        print("PrintPlugin: Hello from the plugin!")


class MathPlugin(Plugin):
    def run(self):
        result = 5 + 10
        print(f"MathPlugin: 5 + 10 = {result}")


class ReversePlugin(Plugin):
    def run(self):
        text = "Asmi"
        print(f"ReversePlugin: {text[::-1]}")


# ---------- Plugin Manager ----------
class PluginManager:
    def __init__(self):
        self.plugins = []

    def load_plugin(self, plugin):
        if issubclass(type(plugin), Plugin) or isinstance(plugin, Plugin):
            self.plugins.append(plugin)
        else:
            print("Error: Not a valid plugin")

    def run_all(self):
        for plugin in self.plugins:
            plugin.run()


# ---------- Example Usage ----------
manager = PluginManager()

# create plugin objects
p1 = PrintPlugin()
p2 = MathPlugin()
p3 = ReversePlugin()

# load plugins
manager.load_plugin(p1)
manager.load_plugin(p2)
manager.load_plugin(p3)

# execute all plugins
manager.run_all()
```

```python
from abc import ABC, abstractmethod

# ---------- Base Plugin ----------
class Plugin(ABC):
    @abstractmethod
    def run(self):
        pass

# ---------- Example Plugins ----------
class PrintPlugin(Plugin):
    def run(self):
        print("Hello from PrintPlugin!")

class MathPlugin(Plugin):
    def run(self):
        print("MathPlugin: 5 + 3 =", 5 + 3)

# ---------- Plugin Manager ----------
class PluginManager:
    def __init__(self):
        self.plugins = []

    def load_plugin(self, plugin):
        if isinstance(plugin, Plugin):   # simple check
            self.plugins.append(plugin)
        else:
            print("Error: Not a valid plugin")

    def run_all(self):
        for plugin in self.plugins:
            plugin.run()

# ---------- Usage ----------
manager = PluginManager()
manager.load_plugin(PrintPlugin())
manager.load_plugin(MathPlugin())

manager.run_all()
```

```python
''' Employee Management

Base class Employee with name, salary, department.

Derived classes: Manager, Developer, Intern.

Each subclass overrides a method calculate_bonus().

Add encapsulation to prevent changing salary directly.'''


# ---------- Base Class ----------
class Employee:
    def __init__(self, name, salary, department):
        self.__name = name               # private attribute
        self.__salary = salary           # private attribute
        self.department = department     # public attribute

    # ---------- Encapsulation ----------
    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        print("Error: Salary cannot be changed directly!")

    def get_name(self):
        return self.__name

    # ---------- Bonus Method ----------
    def calculate_bonus(self):
        return 0   # base class default bonus


# ---------- Derived Classes ----------
class Manager(Employee):
    def calculate_bonus(self):
        # Managers get 20% bonus
        return self.get_salary() * 0.2

class Developer(Employee):
    def calculate_bonus(self):
        # Developers get 15% bonus
        return self.get_salary() * 0.15

class Intern(Employee):
    def calculate_bonus(self):
        # Interns get 5% bonus
        return self.get_salary() * 0.05


# ---------- Example Usage ----------
emp1 = Manager("Alice", 100000, "HR")
emp2 = Developer("Bob", 80000, "IT")
emp3 = Intern("Charlie", 30000, "Marketing")

employees = [emp1, emp2, emp3]

for emp in employees:
    print(f"Name: {emp.get_name()}, Department: {emp.department}")
    print(f"Salary: {emp.get_salary()}, Bonus: {emp.calculate_bonus()}")
    print("-"*40)

# Trying to change salary directly
emp1.set_salary(200000)   # Error: cannot change
```

```python
'''Library Management System

Base class Book with title, author, isbn.

Derived classes: EBook (with download link), PrintedBook (with shelf location).

Library class maintains a collection of books and supports search, borrow, return.

Implement encapsulation for private attributes like borrower_info.'''



class Book:
    def __init__(self,title,author,isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.__borrowerinfo = None

    def get_borrower(self):
        return self.__borrowerinfo
    def set_borrower(self,borrower_name):
        if self.__borrowerinfo is None:
            self.__borrowerinfo = borrower_name
        else:
            print(f"Error: '{self.title}' is already borrowed by {self.__borrower_info}.")
                  
    def return_book(self):
        if self.__borrower_info:
            print(f"'{self.title}' returned by {self.__borrower_info}.")
            self.__borrower_info = None
        else:
            print(f"'{self.title}' is not borrowed by anyone.")

# ---------- Derived Classes ----------
class EBook(Book):
    def __init__(self, title, author, isbn, download_link):
        super().__init__(title, author, isbn)
        self.download_link = download_link

class PrintedBook(Book):
    def __init__(self, title, author, isbn, shelf_location):
        super().__init__(title, author, isbn)
        self.shelf_location = shelf_location


# ---------- Library Class ----------
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def search_by_title(self, title):
        return [book for book in self.books if book.title.lower() == title.lower()]

    def borrow_book(self, title, borrower_name):
        found_books = self.search_by_title(title)
        if found_books:
            book = found_books[0]
            book.set_borrower(borrower_name)
        else:
            print(f"Book '{title}' not found.")

    def return_book(self, title):
        found_books = self.search_by_title(title)
        if found_books:
            book = found_books[0]
            book.return_book()
        else:
            print(f"Book '{title}' not found.")


# ---------- Example Usage ----------
library = Library()

# Add books
book1 = PrintedBook("Python Basics", "Asmi Jain", "ISBN001", "Shelf A1")
book2 = EBook("Advanced Python", "Rahul Kumar", "ISBN002", "www.downloadlink.com/advpython")

library.add_book(book1)
library.add_book(book2)

# Borrow a book
library.borrow_book("Python Basics", "Alice")
library.borrow_book("Python Basics", "Bob")  # Already borrowed

# Return a book
library.return_book("Python Basics")
library.return_book("Python Basics")  # Not borrowed

# Check borrower info (encapsulation)
print("Borrower of EBook:", book2.get_borrower())
```

```python
# Parent Class
class Animal:
    def speak(self):
        print("Animals can make sounds")

# Child Class 1
class Dog(Animal):
    def bark(self):
        print("Dog barks üê∂")

# Child Class 2
class Cat(Animal):
    def meow(self):
        print("Cat meows üê±")

# Child Class 3
class Cow(Animal):
    def moo(self):
        print("Cow moos üêÑ")


# --- Create Objects of Each Child Class ---
dog1 = Dog()
cat1 = Cat()
cow1 = Cow()

# --- All Children can access parent method ---
dog1.speak()   # from Animal class
cat1.speak()
cow1.speak()

# --- Each child has its own specific behavior ---
dog1.bark()
cat1.meow()
cow1.moo()
```
